```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Mean values
mu_values = [1,2,4,10]

# Define a range for x values
x = np.arange(0, 20)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot each Poisson distribution
for mu in mu_values:
    y = poisson.pmf(x, mu)
    plt.plot(x, y, marker='o', label=f'$\mu = {mu}$')

# Adding labels and title
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Poisson Distribution for different $\mu$ values')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
